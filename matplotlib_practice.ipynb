```python
# Import dependencies. 

import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot

plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()

fig = plt.figure()

ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
```

```python
# Create the plot and add a lael for the legend.
plt.plot(x_axis, y_axis, label='Boston', marker="*", color="blue", linewidth=2)

# Create lables for the x and y axes 

plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45 

plt.ylim(0, 45)

# Create a title

plt.title("PyBer Fare by Month")

# Add the legend

plt.legend()
```

```python
fig = plt.figure()

ax = fig.add_subplot()
ax.plot(x_axis, y_axis, color="green", marker="d", label='Boston', linewidth=2)
ax.set_xlabel('Date')
ax.set_ylabel('Fare($)')
ax.set_ylim(0, 45)
ax.legend()
ax.grid()
```
